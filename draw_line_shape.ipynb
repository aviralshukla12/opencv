```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the empty canvas
height , width = 500 , 700
blue = (255,127,0)
canvas =   np.full((height , width ,3), blue , dtype = np.uint8)
plt.imshow(cv2.cvtColor(canvas , cv2.COLOR_BGR2RGB))
```

DRAW TO A LINE

```python
cv2.line(canvas  , (50 ,50) ,(650,450) , (255,255, 255),5)
plt.imshow(cv2.cvtColor(canvas , cv2.COLOR_BGR2RGB))
           
```

DRAW A RECTANGLE

```python
cv2.rectangle(canvas , (100 ,100), (500 , 300) , (0,255,0) , 3)
plt.imshow(cv2.cvtColor(canvas , cv2.COLOR_BGR2RGB))
```

DRAW A CIRCLE

```python
cv2.circle(canvas , (500 , 150) , 100 ,(0,0,255) , 5  )
plt.imshow(cv2.cvtColor(canvas , cv2.COLOR_BGR2RGB))
```

DRAW A ELLIPSE

```python
cv2.ellipse(canvas , (350 ,350) ,(100,50) , 45, 0 , 360 , (255,0,0) ,3)
plt.imshow(cv2.cvtColor(canvas , cv2.COLOR_BGR2RGB))
```

DRAW A POLYGON

```python
# Xn, Yn where n > 2

points = np.array([[150,400] , [300,400] ,[250, 300]] , np.int32)
new_reshaped_points = points.reshape((-1,1,2))

print(new_reshaped_points)

cv2.polylines(canvas ,[new_reshaped_points] , isClosed = True , color = (0,0,0) , thickness = 3)

plt.imshow(cv2.cvtColor(canvas , cv2.COLOR_BGR2RGB))
```
