```python
import cv2

# Open input video
video = cv2.VideoCapture("./mountain.mp4")

if not video.isOpened():
    print("Error: Cannot open video file.")
    exit()

# Get width and height of the video
width  = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
#fps    = video.get(cv2.CAP_PROP_FPS)

# Define codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter("./output_video.mp4", fourcc, 30, (width, height))

while True:
    ret, frame = video.read()
    
    if not ret:
        print("Video processing completed.")
        break

    # Write frame to output video
    frame = frame + 50  # Example processing: increase brightness
    out.write(frame)

    # Display frame
    cv2.imshow("Video Window", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
out.release()
video.release()
cv2.destroyAllWindows()
```

```python
import cv2
bgr_img = cv2.imread("./mountain.jpg")
```

```python
# Convert BGR to Grayscale
gray_img = cv2.cvtColor(bgr_img , cv2.COLOR_BGR2GRAY)
cv2.imwrite("gray_image.jpg", gray_img)
```

```python
#convert to HSV
hsv_img = cv2.cvtColor(bgr_img , cv2.COLOR_BGR2HSV)
cv2.imwrite("./hsv_image.jpg", hsv_img)
```

```python
#convert to lab
lab_img = cv2.cvtColor(bgr_img , cv2.COLOR_BGR2Lab)
cv2.imwrite("./lab_image.jpg", lab_img)
```

```python
# convert to RGB
rgb_img = cv2.cvtColor(bgr_img , cv2.COLOR_BGR2RGB)
cv2.imwrite("./rgb_image.jpg", rgb_img)
```
