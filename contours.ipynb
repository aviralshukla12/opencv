```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./images/shapes.jpg')

image_rgb =  cv2.cvtColor(image , cv2.COLOR_BGR2RGB)

image_gray  =cv2.cvtColor(image_rgb  ,  cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(image_rgb)
```

```python
gray = cv2.Canny(image_gray , 50 , 150)
plt.imshow(gray , cmap='gray')
```

```python
contours  ,hierarchy =  cv2.findContours(gray, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
#MODE  : HIERARCHY RELATIONSHIP BETWEEN CONTOURS
#METOHD : CONTOURS POINTS ARE STORED

contours_image = image.copy()
cv2.drawContours(contours_image , contours , -1 , (0,255,0 ), 5)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(gray , cmap='gray')
plt.title("MAsked Image")

plt.subplot(1,3,3)
plt.imshow(contours_image)
plt.title("Contours")
```

```python
retval  ,binary =  cv2.threshold(image_gray , 200,  255 ,cv2.THRESH_BINARY)
plt.imshow(binary ,cmap='gray')

contours ,hierarchy = cv2.findContours(binary , cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE)
#mode : hierarchy relationship between contours
# methods :contours points are stored
contours_image = image.copy()
cv2.drawContours(contours_image , contours , -1 , (0,255,0) , 5)    

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.subplot(1,3,2)
plt.imshow(binary , cmap='gray')
plt.title("Masked Image")
plt.subplot(1,3,3)
plt.imshow(contours_image)
plt.title("Contours")
plt.show()

```
