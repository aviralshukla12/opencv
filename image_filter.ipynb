```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image =  cv2.imread("./images/filter.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(image  , cmap = "gray")
```

```python
x  = -10
horizontal_kernal = np.array([[-1 ,-1 ,-1] , [x ,x,x] , [1,1,1]])
vertical_kernal =  np.array([[-1 ,x,1] , [-1 ,x,1] , [-1 ,x,1]])
diagonal_kernal =  np.array([[x, -1 ,1] , [-1, x , 1] , [1 ,1,  x]])

horizontal_filtered_image = cv2.filter2D(image , -1, horizontal_kernal)
vertical_filtered_image = cv2.filter2D(image , -1, vertical_kernal)
diagonal_filtered_image = cv2.filter2D(image , -1, diagonal_kernal)

plt.figure(figsize = (10,10))
plt.subplot(1,3,1)
plt.imshow(horizontal_filtered_image , cmap = "gray")
plt.title("Horizontal Filtered Image")  
plt.subplot(1,3,2)
plt.imshow(vertical_filtered_image , cmap = "gray")
plt.title("Vertical Filtered Image")

plt.subplot(1,3,3)
plt.imshow(diagonal_filtered_image , cmap = "gray")

plt.title("Diagonal Filtered Image")

plt.show()
```
