```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./images/giraffe-Kenya.png")
image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
```

FLIP 

```python
# 1 ---> HORIZONTAL FLIP
# 0 ---> vERTICALLY FLIP
# -1 ---> BOTH DIRECTION FLIP
flip_horizontal = cv2.flip(image , 1)
flip_vertical = cv2.flip(image , 0)
flip_both = cv2.flip(image , -1)

titles  = ["Original imgage" , "Horizontal Flip" , "Vertical Flip" , "Flip both axis"]
images = [image_rgb, 
          cv2.cvtColor(flip_horizontal , cv2.COLOR_BGR2RGB) , 
          cv2.cvtColor(flip_vertical , cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_both ,  cv2.COLOR_BGR2RGB)]
plt.figure(figsize = (12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

ROTATE

```python
rotate_90_clockwise = cv2.rotate(image_rgb ,  cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(image_rgb ,  cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image_rgb ,  cv2.ROTATE_180)
titles  = ["Original imgage" , "Rotate 90 Clockwise" , "Rotate 90 CounterClockwise" , "Rotate 180"]
images = [image_rgb,
            rotate_90_clockwise , 
            rotate_90_counterclockwise,
            rotate_180]
plt.figure(figsize = (12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

CROPING IMAGE

```python
print(image_rgb.shape)
```

```python
x_start ,  y_start  = 400, 300
x_end , y_end = 800, 900

cropped_image = image_rgb[y_start: y_end , x_start : x_end]

plt.figure(figsize = (8,8))
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("Off")

plt.subplot(1,2,2)
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis("Off")

plt.tight_layout()
plt.show()
```
