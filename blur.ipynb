```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./images/giraffe-Kenya.png")
image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
kernel_s = 10
average_blur =  cv2.blur(image_grey , (kernel_s , kernel_s))
plt.imshow(average_blur ,  cmap  = 'grey')
```

```python
kernel_s = 31
guassain_blur =  cv2.GaussianBlur(image_grey , (kernel_s , kernel_s) ,0)
plt.imshow(guassain_blur ,  cmap  = 'grey')
```

```python
median_blur = cv2.medianBlur(image_grey , 11)
plt.imshow(median_blur ,  cmap  = 'grey')
```

```python
plt.figure(figsize=(14, 12))

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.subplot(2, 2, 2)
plt.imshow(average_blur ,  cmap  = 'grey')
plt.title("Average Blur")
plt.subplot(2, 2, 3)
plt.imshow(guassain_blur ,  cmap  = 'grey')
plt.title("Gaussian Blur")
plt.subplot(2, 2, 4)
plt.imshow(median_blur ,  cmap  = 'grey')
plt.title("Median Blur")
plt.show()
```
