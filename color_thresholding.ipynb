```python
print(os.getcwd())
#Build the path to the image(make sure that images folder do exist here)
image_path   =  os.path.join(os.getcwd(), 'images' , "giraffe-Kenya.png")
```

```python
image = cv2.imread("./images/giraffe-Kenya.png")
```

```python
if image is None:
    print("Error: Image not found")
else :
    print("Image loaded successsfully")
```

```python
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
image_hsv = cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
plt.imshow(image_hsv)
```

```python
lower_white = np.array([0,0,190])
upper_white =  np.array([30,80,255])

mask = cv2.inRange(image_hsv , lower_white , upper_white)
```

```python
plt.imshow(mask)
```

```python
black_hsv = np.uint8([[[120,255,0]]])
black_bgr = cv2.cvtColor(black_hsv , cv2.COLOR_HSV2BGR)[0][0]
```

```python
image_result = image.copy()
image_result[mask  > 0 ] = black_bgr
```

```python
image_result_rgb = cv2.cvtColor(image_result  , cv2.COLOR_BGR2RGB)
```

```python
#PLOTTING THE IMAGES SIDE BY SIDE

fig , axs = plt.subplots(1,3, figsize =(12,6))

axs[0].imshow(image_rgb)
axs[0].set_title("Original Image")
axs[0].axis('off')

axs[1].imshow(mask)
axs[1].set_title("Binary Mask")
axs[1].axis('off')

axs[2].imshow(image_result_rgb)
axs[2].set_title("Result Image")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
